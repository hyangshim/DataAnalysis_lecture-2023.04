```python
import seaborn as sns
titanic =sns.load_dataset('titanic')
# titanic.dropna(axis=0,inplace=True)
titanic.head()
```

```python
# 결측치 하나씩 제거 및 대체
titanic.drop(columns=['deck'],inplace=True)
titanic.embarked.fillna('S',inplace=True)  #결측치가 별로 없어서 최빈값으로 대체
titanic.embarked.fillna('Southampthon',inplace=True)
```

```python
titanic.isna().sum().sum()
```

```python
titanic['age']=titanic['age'].fillna(titanic.age.median(),inplace=True)
titanic
```

```python
titanic['cat']=pd.qcut(titanic.age,3,labels=['A1','A2','A3'])
titanic.head()
```

```python
titanic.groupby(['sex','embark_town'])['survived'].mean()
```
