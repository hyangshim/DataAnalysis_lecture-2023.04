### 3.인구소멸 위기 지역 top 10
    - 가임기 여성인구수(20-39) / 60세 이상 노령인구
    - 특례시 : 수원 고양, 용인
    - 행정구 : 성남,안산,안양

```python
# 1) 데이터 파일 읽어오기
pail='data/주민등록인구집계현황 (1).csv'
age=pd.read_csv(pail,encoding='euc-kr')
age.head(5)
```

```python
age.isna().sum().sum()
```

```python
age.columns,age.values
```

```python
age.drop(columns=['연도', '월', '행정구역구분명', '총 인구수', '0~9세', '10~19세', '20~29세',
        '30~39세', '40~49세', '50~59세', '총 인구수 (남)', '0~9세 (남)', '10~19세 (남)', '20~29세 (남)',
        '30~39세 (남)', '40~49세 (남)', '50~59세 (남)', '60~69세 (남)', '70~79세 (남)',
        '80~89세 (남)', '90~99세 (남)', '100세 이상 (남)', '총 인구수 (여)', '0~9세 (여)',
        '10~19세 (여)',  '40~49세 (여)', '50~59세 (여)',
        '60~69세 (여)', '70~79세 (여)', '80~89세 (여)', '90~99세 (여)', '100세 이상 (여)'],inplace=True)
age.head(2)
```

```python
for column in age.columns[1:]:
    age[column] = age[column].astype(int)
age.head(3)
```

```python
age['가임기여성인구수'] = age['20~29세 (여)']+age ['30~39세 (여)']
age['60세이상노령인구'] =age['60~69세']+age['70~79세']+age['80~89세']+age['90~99세']+age['100세 이상']
age
```

```python
age.columns
```

```python
age=age[['행정구역명','가임기여성인구수','60세이상노령인구']]
age.head(2)
```

```python
age['인구소멸비율'] = (age['가임기여성인구수'] / age['60세이상노령인구']).round(2)
age.head(5)
```
