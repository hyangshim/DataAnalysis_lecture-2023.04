```python
import numpy as np
import pandas as pd
```

```python
pail='data/서울시 자치구 년도별 CCTV 설치 현황_221231기준 copy.csv'
cctv=pd.read_csv(pail,encoding='euc-kr')
cctv.head()
```

```python
#인덱스 계 삭제함
cctv.drop(index=0,inplace=True)
cctv
```

```python
cctv.index,cctv.columns,cctv.values
```

```python
columns = '구분,2013년 이전\r\n설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
for col in columns[1:]:
    cctv[col] = cctv[col].str.replace(',','').astype(int)
cctv.head()
```

```python
df1 = cctv.loc[:, ['2022년', '2021년', '2020년']].sum(axis=1)
df1
```

```python
df2=cctv[['2013년 이전\r\n설치된 CCTV', '2013년', '2014년', '2015년', '2016년',
        '2017년', '2018년', '2019년']].sum(axis=1)
df2
```

```python
cctv['최근 증가율'] = round((df1 / df2)*100,2)
cctv.head(2)
```

```python
cctv.drop(columns=['2013년 이전\r\n설치된 CCTV', '2013년', '2014년', '2015년', '2016년',
        '2017년', '2018년', '2019년', '2020년', '2021년', '2022년'],inplace=True)
cctv.head(2)
```

```python
cctv.columns=['구별','CCTV댓수','최근증가율']
cctv.head(5)
```

```python
cctv.to_csv('cctv.csv')
```
