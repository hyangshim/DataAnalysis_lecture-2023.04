```python
import numpy as np
import pandas as pd
```

```python
pail='data/서울시인구.txt'
person=pd.read_csv(pail, sep='\t', thousands=',')
person.head(5)
```

```python
person.drop(index=[0,1,2],inplace=True)
person
```

```python
person.drop(columns=['기간','세대', '인구.1', '인구.2', '인구.4', '인구.5',
        '인구.7', '인구.8', '세대당인구'],inplace=True)
person.head(5)
```

```python
person.columns=['구별','인구수','내국인','외국인','고령자']
person.head(5)
```

```python
person.index,person.columns,person.dtypes
```

```python
columns = '구별,인구수,내국인,외국인,고령자'.split(',')
for col in columns[1:]:
    person[col] = person[col].str.replace(',','').astype(int)
person.head(5)
```

```python
person['외국인비율']=round((person['외국인']/person['인구수']) *100,2)
person.head(5)
```

```python
person['고령자비율'] = round((person['고령자']/person['인구수']) *100,2)
person.head(5)
```

```python
person.to_csv('pop.csv')
```
