### 8.합성

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# inner join(교집합) - 중복된 열 기준으로, 공통된 데이터만 합성
pd.merge(df1,df2)
```

```python
# full outer join
pd.merge(df1,df2,how='outer')
```

```python
# Left outer join
pd.merge(df1,df2,how='left')     # 파일 순서에 따라 왼 오 정함
```

```python
#right outer join
pd.merge(df1,df2,how='right')        # 오른쪽 데이터가 나옴
```

- join method

```python
#index로 조인
df1.set_index('고객번호',inplace=True)
df2.set_index('고객번호',inplace=True)

```

```python
# 왼쪽 파일에 + 중복된 데이터 (Left outer join)
df1.join(df2)
```

```python
# 중복되는 데이터는 추출
pd.merge(df1,df2,left_index=True,right_index=True)
```
