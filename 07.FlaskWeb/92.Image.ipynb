```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('static/img/dead-lift.jpg')
plt.imshow(img), plt.axis('off');
```

```python
np.array(img).shape
```

```python
h, w, _ = np.array(img).shape
```

```python
if h > w:
    width, height = w, w
else:
    width, height = h, h
```

```python
diff = abs(h - w) // 2
```

```python
if w > h:           # landscape image
    final_img = np.array(img)[:, diff:diff+h, :]
else:               # portrait image
    final_img = np.array(img)[diff:diff+w, :, :]
```

```python
new_img = Image.fromarray(final_img)
plt.imshow(new_img);
```
