### YouTube Ranking 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/20230425_YouTube랭킹.csv')
df.head(3)
```

- 구독자수 Top 20 - barh 차트로

```python
plt.figure(figsize=(12,8))
plt.barh(df.채널명.head(20), df.구독자수.head(20))
plt.title('구독자수 Top 20 채널')
plt.show()
```

```python
df1 = df.head(20).sort_values(by='구독자수')
plt.figure(figsize=(12,7))
plt.barh(df1.채널명, df1.구독자수)
plt.title('구독자수 Top 20 채널')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(y='채널명', x='구독자수', data=df.head(20))
plt.title('구독자수 Top 20 채널')
plt.show()
```

- 조회수 Top 20

```python
df1 = df.sort_values(by='조회수', ascending=False)

plt.figure(figsize=(12,7))
sns.barplot(y='채널명', x='조회수', data=df1.head(20))
plt.title('조회수 Top 20 채널')
plt.show()
```

- 비디오 수 Top 20

```python
df2 = df.sort_values(by='비디오', ascending=False)

plt.figure(figsize=(12,7))
sns.barplot(y='채널명', x='비디오', data=df2.head(20))
plt.title('비디오수 Top 20 채널')
plt.show()
```

- 카테고리별 채널수

```python
df3 = df.카테고리.value_counts().to_frame()
df3.head(10)
```

```python
df.pivot_table('채널명', '카테고리', aggfunc='count').sort_values('채널명', ascending=False).head(10)
```

```python
sns.barplot(y=df3.index[:10], x='count', data=df3.head(10))
plt.title('카테고리별 채널수 Top 10');
```

- 카테고리별 구독자수 Top 10

```python
df4 = df.pivot_table('구독자수', '카테고리', aggfunc='sum').sort_values('구독자수', ascending=False)
df4.head(10)
```

```python
sns.barplot(y=df4.index[:10], x=df4.구독자수.head(10))
plt.title('카테고리별 구독자수 Top 10');
```
