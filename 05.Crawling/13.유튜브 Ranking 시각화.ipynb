### 유튜브 Ranking 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df=pd.read_csv('data/유튜브 랭킹.csv')
df.head(3)
```

- 구독자수 Top  20 -barh 차트로

```python
df1=df.loc[:19]
df1.tail()
```

```python
plt.barh(df1.채널명,df1.구독자수)
plt.title('구독자수 Top  20 -barh 차트로')
plt.show()
```

```python
df2 =df1.sort_values(by='조회수',ascending=False)
plt.figure(figsize=(12,8))
plt.barh(df2.채널명,df2.조회수)
plt.title('조회수 Top  20 -barh 차트로')
plt.show()
```

```python
df3 =df1.sort_values(by='비디오',ascending=False)
plt.barh(df3.채널명,df3.비디오)
plt.title('비디오 Top  20 -barh 차트로')
plt.show()
```

```python
#카테고리로 카운트
df5 = df['카테고리'].value_counts().head(10)
df5 = pd.DataFrame(df5)
df5
```

```python
df.pivot_table('채널명','카테고리',aggfunc='count').sort_values('채널명',ascending=False).head(10)
```

```python
sns.barplot(y=df5.index,x='count',data =df5);
```

- 카테고리별 구독자수 TOP 10

```python
df6 =df.pivot_table('구독자수','카테고리',aggfunc='sum').sort_values(by='구독자수',ascending=False)
df6.head(10)
```

```python
# x와 y값만 줘도됨 숫자는 10위까지 보여달라는 뜻
sns.barplot(y=df6.index[:10], x=df6.구독자수.head(10));
```
