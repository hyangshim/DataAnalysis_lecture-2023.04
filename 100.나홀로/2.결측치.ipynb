```python
import pandas as pd
import numpy as np
```

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
data
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data)
df
```

```python
# 데이터 채우기 fillna
df.fillna('')  # NaN 테이터를 빈칸으로 채움
```

```python
df['학교'] = np.nan # 학교 데이터 전체를 NaN으로 채움
df
```

```python
df.fillna('모름')
```

```python
df
```

```python
df.fillna('모름',inplace = True)  #수정한 값이 저장됨
```

```python
df
```

```python
df = pd.DataFrame(data)
df
```

```python
df.['SW특기'].fillna('확인중') # SW특기 데이터 중에서 NaN에 대해서 채움
```

```python
# 테이터 제외하기 dropna
df.dropna() #전체 데이터 중에서 NaN 을 포함하는 데이터 삭제
```

```python
# axis :index or columns
# how: any or all
df.dropna(axis ='index',how='any') # nan가 하나라도 있는 row 삭제
```

```python
df.dropna(axis ='columns') # nan가 하나라도 있는 columns 삭제
```

```python
df['학교'] =np.nan
df
```

```python
df.dropna(axis='columns',how='all') # 데이터 전체가 NaN인 경우에만 columns 삭제
```
