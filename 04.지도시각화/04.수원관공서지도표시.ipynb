### 수원 소재 관공서 지도위에 표시하기

```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('data/수원소재관공서.csv')
df['color'] = ['red','green','blue','orange','pink']
df
```

```python
map = folium.Map(location=[df.위도.mean(),df.경도.mean()], zoom_start=12)
for i in df.index:
    folium.Circle(
        location=[df.위도[i],df.경도[i]],
        popup=folium.Popup(df.주소[i],max_width=200),
        tooltip=df.이름[i],
        radius=300,color=df.color[i],fill=True
    ).add_to(map)
title_html ='<h3 align="center" style ="font-size:20px">수원 소재 주요 관광서</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
