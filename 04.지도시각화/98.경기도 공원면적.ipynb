```python
import numpy as np
import json,folium
import pandas as pd
import folium
```

```python
df = pd.read_csv('data/경기도 도시공원정보표준데이터.csv', encoding='euc-kr')
df.info()
```

```python
df = df[['소재지지번주소','공원면적']]
df.columns =['주소','공원면적']
df.head(10)
```

```python
df.dropna(axis=0,inplace=True)
df.head()
```

```python
df=df[df.주소.str.contains('경기도')]
df.head()
```

```python
my_list = []

for name in df.주소:
    if name.find('구 ') >=0:
        my_list.append(''.join(name.split()[1:3]))
    else:
        my_list.append(''.join(name.split()[1]))
df['시군구']=my_list
df
```

```python
df= df.groupby('시군구')['공원면적'].sum().reset_index()
df.head()
```

```python
df.columns=['지역','면적',]
df
```

```python
df = df.set_index('지역')
df  
```

```python
file='data/경기도_geo_simple.json'
geo_data=json.load(open(file,encoding='utf-8'))
```

```python
import numpy as np
def get_text_location(geo_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            gu_dict[gu['properties']['name']] = [np.mean(geo[:,1]), np.mean(geo[:,0])]
    return gu_dict
gu_dict = get_text_location(geo_data)
gu_dict
```

```python
map = folium.Map([37.55,126.98],zoom_start=11,tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_data,
    data=df.면적,               
    columns=[df.index,df.면적],  
    fill_color='YlOrRd',                
    key_on='properties.name'               
).add_to(map)
for gu_name in df.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(map)
title_html = '<h3 align="center" style="font-size:20px">경기도 자치구별 공원 면적</h3>' 
map.get_root().html.add_child(folium.Element(title_html))
map
```

```python
pop = pd.read_csv('../02.Pandas/data/주민등록인구집계현황 (1).csv', encoding='euc-kr')
pop.head()
```

```python
pop.columns
```

```python
pop=pop[['행정구역구분명','행정구역명','총 인구수']]
pop.head()
```

```python
pop[pop.행정구역명.str.contains('안양')]
pop = pop[pop.행정구역구분명.str.contains('시군') | pop.행정구역구분명.str.contains('구')]
pop.head()
```

```python
gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
pop.행정구역명 = pop.행정구역명.str.strip()
drop_index = []
for index in pop[pop.행정구역구분명 == '시군'].index:
    if pop.행정구역명[index].split()[-1] in gu_city:
        drop_index.append(index)
drop_index
```

```python
pop.drop(drop_index, inplace=True)
pop.head()
```

```python
my_list = []

for name in pop.행정구역명:
    if name.find('구') >=0:
        my_list.append(''.join(name.split()[1:3]))
    else:
        my_list.append(''.join(name.split()[1]))
pop['시군구']=my_list
pop.head()
```

```python
pop= pop.groupby('시군구')['총 인구수'].sum().reset_index()
pop.head()
```

```python
pop.columns=['지역','총인구수']
pop.set_index('지역',inplace=True)
pop.head()
```

```python
pop=pop.pivot_table('총인구수','지역',aggfunc='sum')
pop.head()
```

```python
combine = pd.merge(df,pop,left_index=True,right_index=True)
combine.head()
```

```python
combine['인당공원']=(combine.면적 / combine.총인구수  ).round(2)
combine.head()
```
