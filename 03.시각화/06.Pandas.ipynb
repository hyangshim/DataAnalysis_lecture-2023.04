### 5. pansdas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

- line graph

```python
np.random.seed(2023)
df=pd.DataFrame(np.random.randn(100,3),columns=list('ABC'),
                index=pd.date_range('2023-04-01',periods=100)).cumsum()
df.head()
```

```python
df.plot(figsize=(5,3))                   #default 가 line graph
plt.title('pandas 의 plot method 사용예')
plt.xlabel('일자'),plt.ylabel('데이터')
plt.show()
```

- Bar chart

```python
iris =sns.load_dataset('iris')
iris.sepal_length[:30].plot(kind='bar',rot=45,figsize=(6,3))
plt.title('Iris Setosa 품종의 sepal lenght');
```

```python
#barh 버전
iris.sepal_length[:30].plot(kind='barh')
plt.title('Iris Setosa 품종의 sepal lenght');
```

```python
df2 = iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(rot=0,figsize=(6,3))
plt.title('Iris 품종별 Feature의 평균');
```

```python
df2.columns.name ='Feature'
df2.plot.bar(rot=0,figsize=(6,3))
plt.ylim(0,10)         #범례가 그림자 겹치지 않게
plt.title('Iris 품종별 Feature의 평균');
```

- pie chatr

```python
titanic =sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct ='%.1f%%')  #autopct 파이 그래프 글자
plt.title('선실별 승갱수');
```

- histogram, kernel density

```python
iris.plot.hist()
plt.title('Iris 피쳐 Histogram')
iris.plot.kde()     # scipy 모듈 설치
plt.title('Iris 피쳐 kernel density')
plt.show()
```

- box plot

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('Iris Setosa 품종의 각 피쳐별 BOX plot');
```

```python
iris.boxplot(by='species')
plt.tight_layout()
plt.suptitle('Iris 각 피쳐의 품종별 박스 플롯');
```
